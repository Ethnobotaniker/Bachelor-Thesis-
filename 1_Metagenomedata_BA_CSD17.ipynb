```python
project_name = "Metagenomedata_BA_CSD17_Processed" 
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
#%matplotlib inline
#pd.set_option("display.max_columns",None)
```

```python
abundance_all=pd.read_csv("bacteria")
```

```python
abundance_SGP=abundance_all.filter(regex='SGP|species').copy()
abundance_ICN=abundance_all.filter(regex='ICN|species').copy()
abundance_NYC=abundance_all.filter(regex='NYC|species').copy()
abundance_HKG=abundance_all.filter(regex='HKG|species').copy()
abundance_TPE=abundance_all.filter(regex='TPE|species').copy()
abundance_TYO=abundance_all.filter(regex='TYO|species').copy()
abundance_ILR=abundance_all.filter(regex='ILR|species').copy()
abundance_IEV=abundance_all.filter(regex='IEV|species').copy()
abundance_VIE=abundance_all.filter(regex='VIE|species').copy()
```

.median()

```python
abundance_VIE["Sum_VIE"]=abundance_VIE.median(axis=1)
abundance_SGP["Sum_SGP"]=abundance_SGP.median(axis=1)
abundance_ICN["Sum_ICN"]=abundance_ICN.median(axis=1)
abundance_NYC["Sum_NYC"]=abundance_NYC.median(axis=1)
abundance_HKG["Sum_HKG"]=abundance_HKG.median(axis=1)
abundance_TPE["Sum_TPE"]=abundance_TPE.median(axis=1)
abundance_TYO["Sum_TYO"]=abundance_TYO.median(axis=1)
abundance_ILR["Sum_ILR"]=abundance_ILR.median(axis=1)
abundance_IEV["Sum_IEV"]=abundance_IEV.median(axis=1)
```

#### Filtering Top 10

```python
top_10_SGP=abundance_SGP.sort_values("Sum_SGP",ascending=False).head(10)
top_10_ICN=abundance_ICN.sort_values("Sum_ICN",ascending=False).head(10)
top_10_NYC=abundance_NYC.sort_values("Sum_NYC",ascending=False).head(10)
top_10_HKG=abundance_HKG.sort_values("Sum_HKG",ascending=False).head(10)
top_10_TPE=abundance_TPE.sort_values("Sum_TPE",ascending=False).head(10)
top_10_TYO=abundance_TYO.sort_values("Sum_TYO",ascending=False).head(10)
top_10_ILR=abundance_ILR.sort_values("Sum_ILR",ascending=False).head(10)
top_10_IEV=abundance_IEV.sort_values("Sum_IEV",ascending=False).head(10)
top_10_VIE=abundance_VIE.sort_values("Sum_VIE",ascending=False).head(10)
```
